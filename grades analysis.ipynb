先删除NAN大于50%的列

```python
for index in grades.columns:
    if grades.isna().sum()[index] >= grades.shape[0]/2:
        del grades[index]
```

```python
grades
```

发现有些行的总成绩为NAN，删除这些数据

```python
grades = grades.dropna()
```

```python
grades
```

可以看到删除了25条数据

还可以看到总成绩里有汉字，下面对这些成绩进行替换

```python
grades['总成绩'].replace({'及格':str(60),'不及格':str(0),'不通过':str(0),'优':str(90),'良':str(80),'中':str(70)},inplace=True)

# 这行代码刚开始会报A value is trying to be set on a copy of a slice from a DataFrame,加了个列索引就不报了？
```

```python
grades
```

根据总成绩进行分类看替换是否成功

```python
grade_group = grades['总成绩'].groupby(grades['总成绩'])
```

```python
grade_group.describe()
```

总共96种成绩，看来是替换成功了

然后根据学生姓名对每人的成绩进行汇总

```python
everyone = grades['总成绩'].groupby(grades['姓名'])
```

```python
everyone.describe()
```

```python
everyone.mean()   # 求每人各科成绩的平均分还是会报错
```
