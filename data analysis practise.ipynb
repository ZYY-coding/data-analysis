对上面随机生成的一组数据进行清洗和填补

首先清洗重复的行

```python
df_4 = df_4.drop_duplicates()
```

```python
df_4
```

然后删除有一半及以上的列都是NAN的行
下面的代码中，shape[1]是该DataFrame的总列数，shape[0]是行数

```python
for index,row in df_4.iterrows():    
    if df_4.isna().sum(axis=1)[index] >= df_4.shape[1]/2:
        df_4 = df_4.drop(index)
```

```python
df_4
```

可以看到只有第一行被删除了

再删除那些有一半以上的行都是缺失值的列

```python
for index in df_4.columns:
    if df_4.isna().sum()[index] >= df_4.shape[0]/2:
        del df_4[index]
```

```python
df_4
```
